```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")

def logistic_growth(P0, K, r, t):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

P0 = 10
K_values = [500, 1000, 1500]
r_values = [0.05, 0.1, 0.2]
t_max = 100


t = np.linspace(0, t_max, 200)

plt.figure(figsize=(15, 10))

for K in K_values:
    P = logistic_growth(P0, K, r_values[1], t)
    plt.plot(t, P, label=f'Capacidade de Suporte K={K}')

plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Crescimento Populacional com Diferentes Capacidades de Suporte')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(15, 10))

for r in r_values:
    P = logistic_growth(P0, K_values[1], r, t)
    plt.plot(t, P, label=f'Taxa de Crescimento r={r}')

plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Crescimento Populacional com Diferentes Taxas de Crescimento')
plt.legend()
plt.grid(True)
plt.show()


P0_values = [10, 50, 100]

plt.figure(figsize=(15, 10))

for P0 in P0_values:
    P = logistic_growth(P0, K_values[1], r_values[1], t)
    plt.plot(t, P, label=f'População Inicial P0={P0}')

plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Crescimento Populacional com Diferentes Populações Iniciais')
plt.legend()
plt.grid(True)
plt.show()
```
